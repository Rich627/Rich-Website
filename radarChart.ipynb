```python
import numpy as np
import matplotlib.pyplot as plt


def drew(data, labels, name):
    # 計算角度
    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

    # 開始畫圖
    fig, ax = plt.subplots(figsize=(4, 4), subplot_kw=dict(polar=True))

    # 繪製數據點和連線
    ax.plot(angles, data, 'o-', linewidth=1)
    ax.fill(angles, data, alpha=0.1)

    # 隱藏刻度標籤
    ax.set_xticks([])
    
    # 隱藏徑向刻度標籤
    ax.set_yticklabels([])

    # 將數值顯示在數據點上
    for angle, d in zip(angles, data):
        ax.text(angle, d + 0.45, str(d), ha='center', va='center')

    # 設定雷達圖的外圈線
    ax.set_rmax(5)  # 設定雷達圖的最大值
    ax.grid(True)   # 顯示網格線
    plt.show()
    path = './static/img/' + name + '.png'
    fig.savefig(path)
```

```python
drew([3, 2, 4, 3],["Data Analysis", "Business Intelligence", "SQL", "Data Visualization"], name="Data Analytics")
drew([4, 4, 4, 2], ["Statistics", "Machine Learning", "Deep Learning", "Featirning Engineer"],name="Data Science")
drew([4, 2, 3, 3], ["Python", "Data Structures", "Algorithms", "Tools"], name="Computer Science")
drew([3, 2, 2, 1], ["Cloud Architect / Big Data", "Data Mining", "Data Warehousing", "ETL"], name="Data Engineering")
drew([3, 3.5, 3, 2], ["Data Analtyics", "Data Science", "Computer Science", "Data Engineering"], name="Overall")
```
